From above, we can tell this is a highly skewed data, over 99.8% of samples have the class 'non-fraud', that being said, if we predict every single examples with class 'non-fraud', we could get 99.8% accuracy, apparently accuracy is not error metrics we are interested in. We should plot the confusion metrix and identify the recall (true positive rate) and balance with the precision, in other words, for those are truely positive(fraud), we like to predict most of them as positive(high true positive rate), but meanwhile, for every example we predict positive, we do not want to have too many false positive either.


Since this is an extreme imbalanced sample, we will make a subsample that has 50%-50% of positive(fraud, 1) and negative(non-fraud, 0), use the subsample to train our model instead of using the whole sample

```python
# check the number of positive samples
fraud['Class'].value_counts()
```

```python
# randomize the order of samples
fraud = fraud.sample(frac=1, random_state=1009)
non_fraud_sub = fraud[fraud['Class'] == 0].iloc[:492]
fraud_sub = fraud[fraud['Class'] == 1]
sub_sample = pd.concat([non_fraud_sub, fraud_sub])

# randomize the order of sub_sample
sub_sample = sub_sample.sample(frac=1)
```

Our first approach is to use different ML models without doing any feature selection. Before fitting our model, we will define a function that implements gridsearch to find the best accuracy as well as time it takes to fit and predict to come up with a creteria of good model

```python
from sklearn.model_selection import GridSearchCV
def get_best_model_accuracy(model, params, X, y):
    grid = GridSearchCV(model, params, error_score=0., cv=5)
    grid.fit(X, y)
    # our classical metric for performance
    print("Best Accuracy: {}".format(grid.best_score_))
    # the best parameters that caused the best accuracy
    print("Best Parameters: {}".format(grid.best_params_))
    # the average time it took a model to fit to the data (in seconds)
    print("Average Time to Fit (s): {}".format(round(grid.cv_results_['mean_fit_time'].mean(), 3)))
    # the average time it took a model to predict out of sample data (in seconds)
    # this metric gives us insight into how this model will perform in real-time analysis
    print("Average Time to Score (s): {}".format(round(grid.cv_results_['mean_score_time'].mean(), 3)))
    return grid.best_params_
   
    
```

```python
# define X and y, since we are using grid search so it is not necessary to split our data into training and test set
X = sub_sample.drop(['Class', 'Time'], axis=1)
y = sub_sample['Class']
```

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
# set up hyperparameters
lgr_params = {'C': [0.01, 0.1, 10, 100], 'penalty': ['l1', 'l2']}
best = get_best_model_accuracy(lgr, lgr_params, X, y)

# use the best parameter to refit the model
lgr = LogisticRegression(**best)
lgr.fit(X, y)
precision = ((lgr.predict(X)==1) & (y==1)).sum()/(lgr.predict(X)==1).sum()
trp = ((lgr.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

```python
# KNN model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn_params = {'n_neighbors': [1,3,5,7,9]}
best = get_best_model_accuracy(knn, knn_params, X, y)

# use the best parameter to refit the model
knn = KNeighborsClassifier(**best)
knn.fit(X, y)
precision = ((knn.predict(X)==1) & (y==1)).sum()/(knn.predict(X)==1).sum()
trp = ((knn.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

```python
# decision tree model
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree_params = {'max_depth':[1,3,5,7], 'criterion': ['gini', 'entropy']}
best = get_best_model_accuracy(tree, tree_params, X, y)

# use the best parameter to refit the model
tree = DecisionTreeClassifier(**best, random_state=1009)
tree.fit(X, y)
precision = ((tree.predict(X)==1) & (y==1)).sum()/(tree.predict(X)==1).sum()
trp = ((tree.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

```python
# random forest model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf_params = {'n_estimators': [10, 50, 100, 500], 'max_depth': [1,3,5,7]}
best = get_best_model_accuracy(rf, rf_params, X, y)

# use the best parameter to refit the model
rf = RandomForestClassifier(**best, random_state=1009)
rf.fit(X, y)
precision = ((rf.predict(X)==1) & (y==1)).sum()/(rf.predict(X)==1).sum()
trp = ((rf.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

```python
# support vector machine model
from sklearn.svm import SVC
svc = SVC()
svc_params = {'C': [0.01, 0.1, 1, 5, 10], 'kernel': ['rbf', 'linear', 'poly']}
best = get_best_model_accuracy(svc, svc_params, X, y)

# use the best parameter to refit the model
svc = SVC(**best, random_state=1009)
svc.fit(X, y)
precision = ((svc.predict(X)==1) & (y==1)).sum()/(svc.predict(X)==1).sum()
trp = ((svc.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

```python
# gradient boosting model
from sklearn.ensemble import GradientBoostingClassifier
gbs = GradientBoostingClassifier()
gbs_params = {'n_estimators': [10, 50, 100, 500], 'max_depth': [2,3,5], 'learning_rate':[0.01, 0.1, 1 ,10],
             'loss': ['deviance', 'exponential']}
best = get_best_model_accuracy(gbs, gbs_params, X, y)

# use the best parameter to refit the model
gbs = GradientBoostingClassifier(**best, random_state=1009)
gbs.fit(X, y)
precision = ((gbs.predict(X)==1) & (y==1)).sum()/(gbs.predict(X)==1).sum()
trp = ((gbs.predict(X)==1) & (y==1)).sum()/(y==1).sum()
print('True positive rate(recall): {}'.format(trp))
print('Precision: {}'.format(precision))
```

It looks like if we want to balance the accuracy and the time it trains the model, decision tree and knn seems to be our best, though random forest and gradient boosting might have perform slightly better in predicting, the overall time it takes is too long. However, one thing to point out is we use only 492 examples out of 280K for our training set, we should do more with different examples to see whether this could generalize before we use it to fit our entire model

```python
# use different sample for model training and compare knn and decision tree
tree_score = {'Accuracy':[], 'True positive rate':[], 'Precision':[]}
knn_score = {'Accuracy':[], 'True positive rate':[], 'Precision':[]}
fraud[fraud['Class'] == 0].sample
for i in range(500):
    # re-sample
    no_fr = fraud[fraud['Class'] == 0].sample(492)
    df = pd.concat([no_fr, fraud_sub])
    
    # transform Amount column
    df['Amount'] = df['Amount']**0.25
    df['Amount'] = scaler.fit_transform(df[['Amount']])
    
    # split X, y for both model
    X_tree = df[p_value['column'].values[:27]]
    X_knn = df[p_value['column'].values[:7]]
    y = df['Class']
    
    # fit the model, using the best parameters generated previously
    tree = DecisionTreeClassifier(criterion='gini', max_depth=5)
    knn = KNeighborsClassifier(n_neighbors=5)
    tree.fit(X_tree, y)
    knn.fit(X_knn, y)
    
    # append scores
    from sklearn.metrics import accuracy_score
    tree_score['Accuracy'].append(accuracy_score(y, tree.predict(X_tree)))
    knn_score['Accuracy'].append(accuracy_score(y, knn.predict(X_knn)))
    tree_score['True positive rate'].append(((tree.predict(X_tree)==1) & (y==1)).sum()/(y==1).sum())
    knn_score['True positive rate'].append(((knn.predict(X_knn)==1) & (y==1)).sum()/(y==1).sum())
    tree_score['Precision'].append(((tree.predict(X_tree)==1) & (y==1)).sum()/(tree.predict(X_tree)==1).sum())
    knn_score['Precision'].append(((knn.predict(X_knn)==1) & (y==1)).sum()/(knn.predict(X_knn)==1).sum())

print('Decision tree')
print('Average accuracy: {}'.format(np.mean(tree_score['Accuracy'])))
print('Average true positive rate(recall): {}'.format(np.mean(tree_score['True positive rate'])))
print('Average precision: {}'.format(np.mean(tree_score['Precision'])))
print('-'*50)
print('KNN')
print('Average accuracy: {}'.format(np.mean(knn_score['Accuracy'])))
print('Average true positive rate(recall): {}'.format(np.mean(knn_score['True positive rate'])))
print('Average precision: {}'.format(np.mean(knn_score['Precision'])))
```

After 500 iterations, we can see that Decision tree model beats the KNN model in every aspect, we will use it for our original model

```python
# fit our original model
tree = DecisionTreeClassifier(criterion='gini', max_depth=5)

# split into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(fraud.drop('Class', axis=1), fraud['Class'], test_size=0.3)

# feature selection
X_train = X_train[p_value['column'].values[:27]]

# feature transformation and scaling
X_train['Amount'] = X_train['Amount']**0.25
X_train['Amount'] = scaler.fit_transform(X_train[['Amount']])

# fit the model
tree.fit(X_train, y_train)
```

```python
# preparing testing set
X_test = X_test[p_value['column'].values[:27]]
X_test['Amount'] = X_test['Amount']**0.25
X_test['Amount'] = scaler.transform(X_test[['Amount']])
print(accuracy_score(y_test, tree.predict(X_test)))
```

```python
# pretty confusion matrix plot
def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.
    
    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix. 
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.
        
    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names, 
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d", cmap="YlGnBu")
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
# plot confusion matrix and show the true positive rate and precision
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, tree.predict(X_test))
print_confusion_matrix(cm, ['no fraud', 'fraud'])
print('True positive rate(recall): {}'.format(((tree.predict(X_test)==1) & (y_test==1)).sum()/(y_test==1).sum()))
print('Precision: {}'.format(((tree.predict(X_test)==1) & (y_test==1)).sum()/(tree.predict(X_test)==1).sum()))
```
