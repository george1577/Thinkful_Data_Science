```python
# loading dataset
player_data = pd.read_csv('data/player_data.csv')
players = pd.read_csv('data/Players.csv')
season_stat = pd.read_csv('data/Seasons_Stats.csv')
```

### choose one variable and plot that variable and plot that variable four different ways

```python
# clean up NaN in players['height'], sns.distplot takes only clean data
player_h = players['height'].dropna()
player_h.head()
```

```python
# plot height in players, histogram
sns.set()
bins = np.arange(160, 240, 2.5)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.distplot(player_h, bins=bins, kde=False)
plt.xlabel('Height(cm)')
plt.ylabel('Count')
plt.title('Histogram', fontsize=18)
plt.show()
```

```python
# plot height in players, boxplot
sns.set()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.boxplot(player_h, showmeans=True, color='cyan')
plt.xlabel('Height(cm)')
plt.show()
```

```python
# plot height in players, kde plot
sns.set()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.distplot(player_h, bins=bins, hist=False)
plt.xlabel('Height(cm)')
plt.ylabel('Density')
plt.title('kde plots', fontsize=18)
plt.show()
```

```python
# plot height in players, kde + hist plot
sns.set()
bins = np.arange(160, 240, 2.5)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.distplot(player_h, bins=bins)
plt.xlabel('Height(cm)')
plt.ylabel('Density')
plt.title('Histogram and kde plot', fontsize=18)
plt.show()
```

### Choose one continuous variable and one categorical variable, and plot them six different ways

```python
# Pick only the unique positions for players, PF, SG, C, SF, PG
season_stat_clean = season_stat[season_stat['Pos'].isin(['PF', 'SG', 'C', 'SF', 'PG'])]
season_stat_clean = season_stat_clean.dropna(subset=['FT%', 'Pos'])
season_stat_clean = season_stat_clean[(season_stat_clean['FT%'] != 0) & (season_stat_clean['FT%'] != 1)]
season_stat_clean['Pos'].value_counts()
```

```python
# Plot barplots for position vs FT%
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.barplot(x='Pos', y='FT%', data=season_stat_clean, ci=None, order=['PF', 'SF', 'C', 'SG', 'PG'])
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% barplot', fontsize=16)
plt.show()
```

```python
# Plot stripplots for position vs FT%
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.stripplot(x='Pos', y='FT%', data=season_stat_clean, order=['PF', 'SF', 'C', 'SG', 'PG'])
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% stripplot', fontsize=16)
plt.show()
```

```python
# Plot boxplot for position vs FT%
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.boxplot(x='Pos', y='FT%', data=season_stat_clean, order=['PF', 'SF', 'C', 'SG', 'PG'], showmeans=True)
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% boxplot', fontsize=16)
plt.show()
```

```python
# Plot violinplot for position vs FT%
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.violinplot(x='Pos', y='FT%', data=season_stat_clean, order=['PF', 'SF', 'C', 'SG', 'PG'])
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% violinplot', fontsize=16)
plt.show()
```

```python
# Plot pointplot for position vs FT%
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.pointplot(x='Pos', y='FT%', data=season_stat_clean, order=['PF', 'SF', 'C', 'SG', 'PG'], join=False)
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% pointplot', fontsize=16)
plt.show()
```

```python
# Plot boxenplot for position vs FT%
sns.set(style="whitegrid")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))
ax = sns.boxenplot(x='Pos', y='FT%', data=season_stat_clean, order=['PF', 'SF', 'C', 'SG', 'PG'])
plt.xlabel('Position')
plt.ylabel('FT%')
plt.title('Positions vs FT% boxenplot', fontsize=16)
plt.show()
```
